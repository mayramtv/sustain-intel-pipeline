* #### Extracts the column names, companies' information and the link of the description page about the company 
Extracts the column names, companies' information and the link to the page about the company from Wikipedia, then from that page extract the link of each company. And create a csv file with that name of the company and the page

```python
# Find the information for each country in the website by quering the text from the table

# Finds the tables in the page 
tablesInPage = soupTopCompanies.find_all("table")

# Selects the second table and finds the headings, rows of data and links inside of the table 
companiesTableNames = [name.get_text().replace("\n", "") for name in tablesInPage[1].find_all("th")]
companiesTableNames.append("wiki link")
companiesTableNames.append("company link")

# Creates a list with the column names to later append each row
companiesText = [companiesTableNames]

# Finds all the rows of the table for each company 
for row in tablesInPage[1].select("tr")[1:]:
    rowText = [company.get_text() for company in row.find_all("td")]
    
    # creates a complete link with the link extracted from the "a" tag
    companyPageLinks = "https://en.wikipedia.org/" + row.find("a")["href"]
    
    # removes new line charcters and commas from the revenue numbers without removing the comas from the city and state. 
    rowText = [cell.replace("\n", "").replace("'","`").replace(", ", "%temp%").replace(",", "").replace("%temp%", "-") for cell in rowText]
    
    # Attaches the link to each row 
    rowText.append(companyPageLinks)
    rowText.append("link")
    
    # Passes each row to the table array
    companiesText.append(rowText)
    
    
print(companiesText[:4])
print(companiesText[34])
```

### 5.3 Scraps Each Websites About Page to Extract All Links
<a id="ID5.3"></a>

* #### Find the links in the **About** page for each company using the respective website. 
I will start by looking directly into the links inside of each company "About" page. Due to time and maybe size constraints, I will narrow the search to this area only and write a message of "Not found" if not found by this method.
When running the following code, I find that there are some pages that don't show links due to those pages being generated dynamically. If the response contains the HTML object then the links are extracted from the object.  

```python
# Extract data from each Wikipedia page

# Contains the objects with the "a" tags
aboutLinks = {}

for company in companiesText[1:]:
    
    # Position of the wiki link
    print(company[1])
    
    # checks if the cell contains a link or just the placeholder "link"
    if company[8] != "link":
        
        headers = {
            'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.9',
        }
    
        # Sends request, loads the webpage content
        try:
            reqCompanyPage = requests.get(company[8]+"about", headers=headers, timeout=5)
            print("Response from the page: ", reqCompanyPage ) 
        except:
            aboutLinks[company[1]] = "page not found"
            continue

        if reqCompanyPage.status_code == 200:

            # Convert to a beautifulSoup object and looks for the "a" tag 
            companyAboutPage = BeautifulSoup(reqCompanyPage.content).find_all("a")

            # Not all companies have an object returned, therefore to avoid having an error the outpt is compared with None.
            if companyAboutPage is not None:
                aboutLinks[company[1]] = companyAboutPage
            else: aboutLinks[company[1]] = "dynamic page"
                
        else: aboutLinks[company[1]] = "not found"
    else: aboutLinks[company[1]] = [] # will keep company in the dictionary with an empty list
        
```

>This is the image of the print statements in the previous cell from the website response of the **about** page. Some websites have a response 404, but in other websites the page was not found and the kernel kept loading, therefore I added a timeout after 5 seconds to stop looking for the page and instead use the string "page not found" in the dictionary value.

```python
# Saves the output to a dictionary with the name of the company as key and the list of links as values
# It checks if the output is bigger than 7 charcters and if it contains the "https" at the beggining of the string

# Contains the links of each company  
companiesLinks = {}

for company in aboutLinks:
    print(company)
    # It checks for errors produced when the "a" tag doesn't contain a "href" link
    try:
        if type(aboutLinks[company]) is not str: 
            hrefLinks = [linkA["href"] for linkA in aboutLinks[company] if len(linkA["href"]) > 7 and linkA["href"][0:5] == "https"]
            print(hrefLinks[:4])
            # Assgns the list of links to the respective company in the dictionary 
            companiesLinks[company] = hrefLinks
        else:
            companiesLinks[company] = [] # will keep company in the dictionary with an empty list
    except:
        companiesLinks[company] = [] # will keep company in the dictionary with an empty list
        continue
```

* #### Send a request to each page where a related word to sustainability appear
By using the **sustainabilityLinks** extracted from each website, a request will be sent to each page to obtain the text in each page. If an error is encountered or the page is not found, the string **"Not found"** will be passed instead of the text object.

```python
# Extract data from each Sustainability page

# Contains the objects with the "a" tags
sustainabilityText = {}

for company in sustainabilityLinks.keys():
    print(company)
    sustList= []
    for sustLink in sustainabilityLinks[company]:
        
        # Sends request, loads the webpage content
        try:
            reqPage = requests.get(sustLink, timeout=5)
            print("Response from the page: ", reqPage ) 
        except:
            continue

        if reqPage.status_code == 200:

            # Convert to a beautifulSoup object and looks for the "a" tag 
            sustainabilitySoup = BeautifulSoup(reqPage.content).get_text()

            # Not all companies have an object returned, therefore to avoid having an error the outpt is compared with None.
            if sustainabilitySoup is not None:
                sustList.append(sustainabilitySoup)
                
        # Checks if the page is forbiden then  adds the company name to a list to remove later
        elif reqPage.status_code == 403: # forbiden pege
            continue
    sustainabilityText[company] = sustList
```

* #### Clean the Text Object  
Once the text object is extracted from each page, then long spaces, new line characters and Merged words will be removed from the string.
Two methods would be used to remove the characters and merge words.  
    * First the string is striped from empty spaces at the beginning and end of a string.
    * Then **re.sub()** is used to remove new line, tabs, long strings of empty spaces, and some character  
    * Finally a pattern is compiled, to be removed from the string

```python
# Example 
print("Kroger", sustainabilityText["Kroger"])
print("Phillips 66", sustainabilityText["Phillips 66"][0][:30])
```

* #### Saves the links by company to a file named "CompaniesLinks.csv"

```python
# Removes new line, tabs, empty spaces, and "\xa0" charcters using "re" library
cleanSustText = {}

for company in sustainabilityText:
    cleanList = []
    for strSoup in sustainabilityText[company]:
        cleanStr = strSoup.strip()
        cleanStr = re.sub("\n+", " ", cleanStr)
        cleanStr = re.sub("\t+", " ", cleanStr)
        cleanStr = re.sub("\xa0", " ", cleanStr)
        cleanStr = re.sub("\s+", " ", cleanStr)
        cleanStr = re.sub("-\n|--|\n-", "", cleanStr)
        
        cleanStr = re.sub(",", "", cleanStr)
        cleanStr = re.sub(":", "", cleanStr)
        cleanStr = re.sub("!", "", cleanStr)
        cleanStr = re.sub("\)", "", cleanStr)
        cleanStr = re.sub("\(", "", cleanStr)
        cleanStr = re.sub("&", "", cleanStr)
        cleanList.append(cleanStr)
    cleanSustText[company] = cleanList
```

```python
# Example with no new line, tabs, empty spaces, and "\xa0" charcters
print("Kroger", cleanSustText["Kroger"])
print("Phillips 66", cleanSustText["Phillips 66"][0][0:400])
```

```python
# Remove merged words. This words are button's names that got extracted when scraping and didn't have a space between them
longWord = re.compile(r'\W*\b\w[a-z]*[A-Z]\w*\b')

for company in cleanSustText:
    cleanSustText[company] = [longWord.sub('', compList) for compList in cleanSustText[company]]
#         clnList = longWord.sub('', compList)
#         clnList = " ".join([word for word in compList.split() if len(compList) > ])
```
